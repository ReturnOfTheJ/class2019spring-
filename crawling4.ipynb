네이버 영화에서 영화인 검색어 top 50을 불러오는 작업이다.

```python
html=urlopen('https://movie.naver.com/movie/sdb/rank/rpeople.nhn')
bsObject=BeautifulSoup(html, 'html.parser')
bsObject
```

위의 작업들을 통해 이름, 생년월일을 불러왔다.

```python
for n in range(0,50):
    td=bsObject.find_all('td','title')[n].get_text()
    birth=bsObject.find_all('td','birth')[n].get_text()
    print(n+1,td, birth)
```

이를 엑셀파일로 변환하는데 성공하였고 엑셀파일은 따로 첨부하였다.

```python
with open ('crawling.csv', 'w') as f:
    writer=csv.writer(f)
    writer.writerow(['rank', 'title', 'birth'])
    
    for n in range(0,50):
        td = bsObject.find_all('td','title')[n].get_text()
        birth = bsObject.find_all('td','birth')[n].get_text()
        writer.writerow([n+1, td, birth])
```
